```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
```

```python
#carrega o da6taset
data = load_breast_cancer()

# seleciona os atributos 
X = data.data
y = data.target

# conevretendo em dataframe
df = pd.DataFrame(X, columns=data.feature_names)
df['target'] = y

df['diagnóstico'] = data.target

traducao = {
    'mean radius': 'raio médio',
    'mean texture': 'textura média',
    'mean perimeter': 'perímetro médio',
    'mean area': 'área média',
    'mean smoothness': 'suavidade média',
    'mean compactness': 'compacidade média',
    'mean concavity': 'concavidade média',
    'mean concave points': 'pontos côncavos médios',
    'mean symmetry': 'simetria média',
    'mean fractal dimension': 'dimensão fractal média',
    
    'radius error': 'erro do raio',
    'texture error': 'erro da textura',
    'perimeter error': 'erro do perímetro',
    'area error': 'erro da área',
    'smoothness error': 'erro da suavidade',
    'compactness error': 'erro da compacidade',
    'concavity error': 'erro da concavidade',
    'concave points error': 'erro dos pontos côncavos',
    'symmetry error': 'erro da simetria',
    'fractal dimension error': 'erro da dimensão fractal',
    
    'worst radius': 'raio pior caso',
    'worst texture': 'textura pior caso',
    'worst perimeter': 'perímetro pior caso',
    'worst area': 'área pior caso',
    'worst smoothness': 'suavidade pior caso',
    'worst compactness': 'compacidade pior caso',
    'worst concavity': 'concavidade pior caso',
    'worst concave points': 'pontos côncavos pior caso',
    'worst symmetry': 'simetria pior caso',
    'worst fractal dimension': 'dimensão fractal pior caso'
}

# Renomear as colunas
df.rename(columns=traducao, inplace=True)

# Exibir as primeiras linhas
print(df.head())


```

```python
X = df.drop('diagnóstico', axis=1)
y = df['diagnóstico']
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC


Xtrain, Xtest, ytrain, ytest =  train_test_split(X, y, random_state = 42) 
print(Xtrain.shape)
print(Xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier


model = DecisionTreeClassifier()
model.fit(Xtrain, ytrain)

importances = pd.Series(model.feature_importances_, index=X.columns)
importances.nlargest(10).plot(kind='barh')
plt.title('Top 10 Atributos Mais Importantes')
plt.show()

```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(Xtest)

print("Acurácia:", accuracy_score(ytest, y_pred))
print(classification_report(ytest, y_pred, target_names=['Maligno', 'Benigno']))
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plot_tree(model, feature_names=X.columns, class_names=['Maligno', 'Benigno'], filled=True)
plt.title("Árvore de Decisão para Diagnóstico de Câncer de Mama")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(Xtrain)
X_test_scaled = scaler.transform(Xtest)
```

```python
from sklearn.svm import SVC

modelo_svm = SVC(kernel='linear', random_state=42)
modelo_svm.fit(X_train_scaled, ytrain)
```

```python
y_pred = modelo_svm.predict(X_test_scaled)

print("Acurácia:", accuracy_score(ytest, y_pred))
print(classification_report(ytest, y_pred, target_names=['Maligno', 'Benigno']))
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

svm = SVC(kernel='linear', C=0.4, decision_function_shape='ovo')
svm.fit(X_pca, y.ravel())

plt.figure(figsize=(10,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=y.ravel(), cmap='coolwarm', edgecolors='k')
plt.title('SVM com PCA (2D)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
# 🔹 Definir hiperparâmetros para testar
param_grid = {
    'n_neighbors': list(range(1, 21)),         # número de vizinhos
    'weights': ['uniform', 'distance'],        # peso dos vizinhos
    'p': [1, 2]                                 # 1 = Manhattan, 2 = Euclidiana
}
```

```python
# 🔹 Aplicar GridSearchCV
grid_knn = GridSearchCV(estimator=knn, param_grid=param_grid, cv=5, n_jobs=-1, verbose=1)
grid_knn.fit(Xtrain, ytrain)
```

```python
# 🔹 Resultados
print("Melhor score (validação cruzada):", grid_knn.best_score_)
print("Melhores parâmetros:", grid_knn.best_params_)
```

```python
# 🔹 Avaliar no conjunto de teste
y_pred = grid_knn.predict(Xtest)
print("\nRelatório de classificação no teste:")
print(classification_report(ytest, y_pred, target_names=['Maligno', 'Benigno']))
```
